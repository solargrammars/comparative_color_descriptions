```python
# just for fun. 
import numpy as np
import matplotlib.pyplot as plt
from scipy import spatial
from sklearn.metrics.pairwise import cosine_similarity
```

```python
plt.figure(figsize=(10,10))
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()
plt.xticks( np.arange(0,10, 1.0))
plt.yticks( np.arange(0,10, 1.0))

V = np.array([[1,4],[6,6] ])   # ref ,tgt
origin = [0], [0] # origin point

tr  = V[1] - V[0] 
w   = [5,4]
w_plus_r = w + V[0]

plt.quiver(*origin, V[0][0], V[0][1], color="r", label= "r", angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, V[1][0], V[1][1], color="b", label= "t", angles='xy', scale_units='xy', scale=1)
plt.quiver(*V[0], tr[0],tr[1],        color="black", label= "tr", angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, w[0], w[1],       color="g", label= "wg",  angles='xy', scale_units='xy', scale=1)
plt.quiver(*V[0], w[0], w[1],         color="g", label="wg",  angles='xy', scale_units='xy', scale=1)

plt.quiver(*origin, w_plus_r[0], w_plus_r[1], color="y", label="r+wg", angles='xy', scale_units='xy', scale=1)
plt.legend()
plt.show()
print("cosine   distance (w,tr): ", spatial.distance.cosine(w,tr))
print("pairwise distance (r+wg, t):", spatial.distance.pdist(
    [ w_plus_r, V[1]  ], "euclidean" ))
```
