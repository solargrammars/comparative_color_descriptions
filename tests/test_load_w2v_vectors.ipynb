```python
 import sys
import ipdb
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint 
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

redu = TSNE(n_components=2)

sys.path.append('../')
from settings import OUTPUT_PATH
from data import load_w2v_sample
```

```python
w2v_vectors = load_w2v_sample()
data = [ (k, v) for k,v in w2v_vectors.items()]
tokens, vectors = list(zip(*data))

vectors_reduced = redu.fit_transform(vectors).tolist()
vrx, vry = zip(*vectors_reduced)

fig, ax = plt.subplots(figsize=(10,10))


for k in range(len(vrx[:30])):
    plt.scatter(vrx[k], vry[k], c="black")

    plt.annotate(tokens[k],
            xy=(vrx[k],vry[k]), xycoords='data',
            xytext=(-10, 10), textcoords='offset points',
            arrowprops=dict(arrowstyle="->"))

plt.show()
```
